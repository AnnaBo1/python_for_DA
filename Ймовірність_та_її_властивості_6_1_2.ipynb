**4 ЗАДАЧА**

Знайти ймовірність того, що при кидку двох гральних кубиків добуток чисел на них:
    
  а) дорівнюватиме семи;
  б) буде не меншим за 20;
  в) буде парним.

    Крок 1 Загальна кількість можливих результатів: 36

Кожен кубик має 6 граней з числами від 1 до 6. Отже, загальна кількість можливих комбінацій при кидку двох кубиків = 6 х 6 = 36

    а) ймовірність, що дорівнюватиме семи: 0

Розглянемо варінти, коли число першого кубика множиться на число другого кубику.

7 може бути отримано тільки якщо 1х7, але 7 не існує на кубику і числа на кубику від 1 до 6, тому ці варіанти неможливі

Отже, жоден варіант не дає добутку 7.

P (добуток дорівнює 7) = 0 / 36 = 0

    б) Ймовірність того, що добуток буде не меншим за 20: 2/9 або 0.2222

Визначимо всі можливі варіанти, де ( а х б > 20 ):

4 х 5 = 20

4 х 6 = 24

5 х 5 = 26

5 х 6 = 30

6 х 6 = 36

5 х 4 = 20

6 х 4 = 24

6 х 5 = 30

Отже, маємо 8 різних комбінацій де добуток чисел будет > 20.

P(добток > 20) = 8 / 36 = 2 / 9 = 0.2222

    в)  ймовірність буде парним: 3 / 4 або 0.75

На кубиках парними є числа: 2, 4, 6.

Тепер знайдемо кількість випадків, коли добуток чисел на кубиках буде непарним, бо ці випадки легше підрахувати . Непарні числа: 1, 3, 5.

1 х 1 = 1

1 х 3 = 3

1 х 5 = 5

3 х 1 = 3

5 х 1 = 5

3 х 3 = 9

3 х 5 = 15

5 х 3 = 15

5 х 5 = 25

Отже, всього є 9 комбінацій, коли добуток непарний.

Тоді кількість комбінацій, коли добуток парний:

N(парних) = 36 - 9 = 27.

Ймовірність того, що добуток буде парним:

N(парний добуток) = 27 / 36 = 3 / 4 = 0.75
