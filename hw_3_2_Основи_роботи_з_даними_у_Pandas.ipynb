9. Врахуовуючи висновок в попередньому завданні, дослідимо як продаються в цих містах різні категорії товарів. Виведіть відсотковий розпоіділ кількості інвойсів по категоріям товарів `Product line` в містах `Yangon` та `Mandalay`.

```python
yangon_df = df[df['City'] == 'Yangon']
mandalay_df = df[df['City'] == 'Mandalay']

y_product_counts = yangon_df['Product line'].value_counts() #підрахунок кількості інвойсів для кожної категорії товарів
y_invoices = len(yangon_df)  #len загальна кількість інвойсів
y_percent = (y_product_counts / y_invoices) * 100 #обчислення відсотків
```

```python
m_product_counts = mandalay_df['Product line'].value_counts()
m_invoices = len(mandalay_df)
m_percent = (m_product_counts / m_invoices) * 100
```

```python
print(y_percent)
print(m_percent)
```
