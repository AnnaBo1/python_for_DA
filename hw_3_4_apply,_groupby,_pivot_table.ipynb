1. Дослідимо, який філіал супермаркету ('Branch') є найприбутковішим. Для цього знайдіть сумарний прибуток за кожним філіалом і виявіть, який філіал має найвищий.

```python
df.head()
```

```python
profit = df.groupby('Branch')['gross income'].sum().reset_index() #без reset_index() видасть помилку. це метод для скидання індексу DataFrame
                                                                  #очислимо сумарний прибуток для кожного філіалу

most_profitable = profit.loc[profit['gross income'].idxmax()]     #.loc[], ми вибираємо рядок за індексом, який отримали раніше (idxmax()).
                                                                  #ми вибираємо рядок з найбільшим прибутком (найприбутковіший філіал).
                                                                  #branch_profit['gross income'].idxmax()
                                                                  #Ця частина коду знаходить індекс (номер рядка) з найбільшим значенням у стовпці 'gross income'. Метод idxmax() повертає індекс рядка, де дохід є максим
most_profitable
profit
```

2. В якому місті знайходиться філіал з найвищим прибутком? Може в тому місці нам розмітисти ще один магазин.  
Знайдіть відповідь за допомогою функціоналу Pandas.

```python
profit = df.groupby(['Branch', 'City'])['gross income'].sum().reset_index()  #групуємо за філіалами і містами, і знаходимо сумарний прибуток по філіалах
most_profitable = profit.loc[profit['gross income'].idxmax()] #шуккаємо філіал з найвищим прибутком
most_profitable
```

```python
city_filter = df[df['City'] == 'Naypyitaw'] #фільтруємо DataFrame за конкретним містом

product = city_filter['Product line'].unique() #унікальні лінійки продуктів у цьому місті

product
```

```python
product_profit = df.groupby(['Product line','City'] )['gross income'].sum().reset_index() #groupby('Product line')['gross income'].sum():
                                                                                          #групуємо дані за лінійками продуктів і підраховуємо сумарний прибуток (gross income) для кожної лінійки.

product_profit_sorted = product_profit.sort_values(by='gross income', ascending=False) #sort_values(by='gross income', ascending=False):
                                                                                       #сортуємо результат у спадному порядку, щоб найбільш прибуткова лінійка була на першому місці.

most_profitable_product_line = product_profit_sorted.iloc[0] # iloc[0]: вибираємо перший рядок із найбільшим прибутком.Виводимо найбільш прибуткову лінійку продуктів

most_profitable_product_line
```

3.1. Створіть зводну таблицю, яка покаже, скільки покупок (інвойсів) було зроблено в кожній з філій (`Branch`) за різними категоріями товарів. Запишіть таблицю в змінну `invoices_by_category` і виведіть змінну на екран.
Ця таблиця допоможе проаналізувати, в якій філії купують найбільше товарів кожної з категорій.

```python
invoices_by_category = df.pivot_table(index='Branch', columns='Product line', values='Invoice ID', aggfunc='count')

invoices_by_category
```

1) Як середній обсяг покупки змінюється в залежності від часу доби?

```python
average_purchase_by_hour = df.groupby('Time')['Total'].mean() #Обчислюємо середнє значення для кожного часу в колонці Total
average_purchase_by_hour
```

```python
average_purchase_by_time = df.groupby('Time')['Total'].mean() #групуємо за колонкою 'Time' і обчислюємо середнє значення в колонці 'Total'
top_average_purchases = average_purchase_by_time.sort_values(ascending=False).head(5) #сортуємо середні значення по спаданню і вибираємо найбільші (наприклад, топ 5)
top_average_purchases
```

Висновки: Час доби може впливати на середній розмір покупки.

3)Чи є кореляція між кількістю покупок і середньою ціною товарів в кожній категорії?

```python
correlation = df.groupby('Product line').agg({'Total': 'count', 'Unit price': 'mean', 'Quantity': 'sum'})
correlation                                #тотал Кількість записів у кожній категорії,Середня ціна товарів у кожній категорії,Сума кількості товарів у кожній категорії
```

```python
correlation = df['Total'].corr(df['Unit price'])
print(f"Кореляція між кількістю покупок і середньою ціною товарів: {correlation:.2f}") #2f округлює число до двох десяткових знаків і відображає його у форматі з фіксованою комою
```

Гіпотеза: Вища ціна товарів може знижувати частоту покупок у певних категоріях.

Висновок: Кореляція 0.63 показує помірний позитивний зв'язок між середньою ціною і кількістю покупок. Це свідчить про те, що в середньому, коли середня ціна товарів зростає, кількість покупок також має тенденцію зростати.
Оскільки існує позитивний зв'язок між середньою ціною і кількістю покупок, можливо, варто переглянути ціни.
