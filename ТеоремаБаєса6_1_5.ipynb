**1 ЗАДАЧА**

У медичному центрі проводять тест на певну хворобу. Відомо, що 1% населення страждає на цю хворобу. Тест дає позитивний результат у 99% випадків для хворих і 5% випадків для здорових людей (хибнопозитивний результат). Якщо тест показав позитивний результат, яка ймовірність того, що людина насправді хвора?

Ймовірність того, що людина хвора Р(хвора) = 0.01

Ймовірність позитивного результату за умови, що людина хвора: Р(позитивний/хвора) = 0.99

Ймовірність позитивного результату за умови, що людина здорова: Р(позитивний/здорова) = 0.05

Ймовірність того, що людина здорова : Р(здорова) = 0.99

Потрібно знайти ймовірність того, що людина насправді хвора, якщо тест показав позитивний результат, тобто Р(позитивний/хвора).


    Зараз знайдемо загалну ймовірність позитивного результату Р(позитивний):

Р(позитивний) = Р(позитивний/хвора) х  Р(хвора) + Р(позитивний/здорова) х  Р(здорова)

Р(позитивний) = (0.99 х 0.01) + (0.05 х 0.99) = 0.0099 + 0.0495 = 0.0594

Отже, ймовірність того, що тест позитивний (незалежно від того, хвора людина чи ні), становить 0.0594 або 5.94%.
