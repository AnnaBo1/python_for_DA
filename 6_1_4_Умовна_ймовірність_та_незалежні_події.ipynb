**1 ЗАДАЧА**

З колоди у 52 карти випадково витягується одна карта. Яка ймовірність того, що це туз, якщо відомо, що ця карта — пікова?

Всього у нас 54 карти і ділимо на кількість мастей 4 = 13 карт кожної масті

Сере цих 13 тільки 1 це піковий туз

1 / 13 = 0.0769

    Відповідь: ймовірність, що витягнута карта — піковий туз, за умови, що вона пікова = 0.0769 або це 7.69 %

**2 ЗАДАЧА**

У групі з 30 студентів є 18 хлопців і 12 дівчат. Відомо, що 7 хлопців і 5 дівчат носять окуляри. Випадково обраний студент носить окуляри. Яка ймовірність того, що цей студент — дівчина?

Всього носять окуляри 7 хлопців + 5дівчат = 12 людей.

Отже, кількість дівчат які носять окуляри треба поділити на загальну кількість людей, які носять окуляри:

5 / 12 = 0.4166

    Відповідь: ймовірність того, що випадково обраний студент, який носить окуляри це дівчина = 0.4166 або 41.66 %

У конверті знаходяться 10 лотерейних білетів, серед яких 3 виграшних. З конверта послідовно витягуються білети. Знайти ймовірності того, що:
    
    а) 2-й витягнутий білет буде виграшним, якщо 1-й був виграшним;
    б) 3-й буде виграшним, якщо попередні два білети були виграшними;
    в) 4-й буде виграшним, якщо попередні білети були виграшними.

    а) Ймовірність того, що 2-й витягнутий білет буде виграшним, якщо 1-й був виграшним: 2 / 9 = 0.2222
    
У нас є 10 білетів, з яких 3 виграшні.

Після того, як 1-й білет був виграшним, залишилося 9 білетів, з яких 2 виграшні.

Ймовірність того, що 2-й білет буде виграшним:

Треба кількість виграшних білетів після 1-го (2) поділити на загальну кількість білетів пілся 1-го витягнутого:

2 / 9 = 0.2222

    б) Ймовірність того, що 3-й витягнутий білет буде виграшним, якщо попередні два білети були виграшними: 1 / 8 = 0.125

Після того, як попередні 2 білети були вигришні, залишилос 8 білетів, з яких 1 виграшний.

Ймовірність того, що 3-й білет буде виграшним:

Треба кількість виграшних білетів після 2-го (2) поділити на загальну кількість білетів пілся 2-го витягнутого:

1 / 8 = 0.125

    в) Ймовірність того, що 4-й витягнутий білет буде виграшним, якщо попередні три білети були виграшними: 0

Якщо всі три попередні білети вже були виграшними, то в конверті залишилося 7 білетів, і всі вони є невиграшними.

Ймовірність того, що 4-й білет буде виграшним:

Кількість білетів після 3-го(0) поділити на загальну кількість білетів після 3-го:

0 / 7 = 0

Оскільки всі виграшні білети вже витягнуті, ймовірність виграшу = 0.
