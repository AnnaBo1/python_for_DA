```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Загрузка данных и предварительная обработка
df = pd.read_csv('Бали 6.0 - Практика_бали.csv').iloc[:-1]  # Удаление последней строки
df.iloc[:, -1] = df.iloc[:, -1].str.replace('%', '').astype(float)  # Удаляем символ % и преобразуем в float
```

```python
# Определение категорий
df['Категория'] = df.iloc[:, -1].apply(lambda score: 'Менее 50%' if score < 50 else 'От 50% до 70%' if score < 70 else '70% и выше')
```

```python
# Подсчет студентов по категориям и расчет процентов
category_counts = df['Категория'].value_counts().reset_index()
category_counts.columns = ['Категория', 'Количество студентов']
category_counts['Процент'] = (category_counts['Количество студентов'] / category_counts['Количество студентов'].sum()) * 100
```

```python
# Вывод результатов
print(category_counts)
```

```python
# Визуализация
plt.figure(figsize=(12, 6))

# Гистограмма
plt.subplot(1, 3, 1)
plt.bar(category_counts['Категория'], category_counts['Количество студентов'], color=['red', 'green', 'yellow'])
plt.title('Количество студентов по категориям')
plt.ylabel('Количество студентов')
plt.xlabel('Категория')
plt.xticks(rotation=45, ha='right') # Поворот подписей оси X
# Установим деления по целым числам на оси Y
plt.yticks(range(0, max(category_counts['Количество студентов']) + 1))  # Устанавливаем метки на оси Y


# Круговая диаграмма
plt.subplot(1, 3, 2)
colors = ['red', 'green', 'yellow']
plt.pie(
    category_counts['Процент'],
    labels=category_counts['Категория'],
    autopct='%1.1f%%',  # Формат отображения процентов
    colors=colors
)
plt.title('Процент студентов по категориям')
plt.axis('equal')

# Столбчатая диаграмма
plt.subplot(1, 3, 3)
plt.bar(category_counts['Категория'], category_counts['Процент'], color=['red', 'green', 'yellow'])
plt.title('Процент студентов по категориям')
plt.ylabel('Процент')
plt.xlabel('Категория')
plt.xticks(rotation=45, ha='right') # Поворот подписей оси X

plt.tight_layout()
plt.show()
```

Total Number of Students: analysis includes 45 students, representing 100% of the total. These students are distributed across three categories based on their performance.

Average Performance Level:

Attention should be given to these students, as they may need additional support or resources to improve their performance.

    High Performance Level:
