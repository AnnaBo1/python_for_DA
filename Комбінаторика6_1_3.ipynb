**1 ЗАДАЧА**

Скільки різних паролів можна скласти, використовуючи три літери англійського алфавіту, якщо літери не повинні повторюватися?

В англ алфавіті = 26 літер.
На перше місце можемо вибрати будь-яку з 26 літер.
На друге місце - вже тільки з 25 літер (одна літера вже використана).
На третє місце - лише 24 літери.

26 х 25 х 24 = 15600

    Відповідь: можна скласти 15 600 різних паролів із трьох літер без повторень.

**3 ЗАДАЧА**

Скількома різними способами можна розташувати на полиці 5 різних книг?

P(n) = n!

n=5

P(5) = 5! = 5 х 4 х 3 х 2 х 1 = 120

    Вілповідь: 5 різних книг можна розташувати на полиці 120 різними способами.

**4 ЗАДАЧА**

В урні знаходяться 4 червоні, 3 сині і 2 зелені кулі. Скількома способами можна витягти 3 кулі, якщо вони повинні бути різних кольорів?

Оскільки ви повинні витягнути 3 кулі різного кольору, потрібно взяти по одній кулі з кожного кольору.

Кількість варіантів вибору для кожного кольору:

Для червоної кулі: є 4 можливості (4 червоні кулі).
Для синьої кулі: є 3 можливості (3 сині кулі).
Для зеленої кулі: є 2 можливості (2 зелені кулі).

4 х 3 х 2 = 24

    Відповідь: можемо витягти 3 кулі різного кольору з урни 24 різними способами.
